```python
numeric_cols = ['wins', 'losses', 'draws', 'height_cm','weight_in_kg', 'reach_in_cm',  'significant_strikes_landed_per_minute',
       'significant_striking_accuracy', 'significant_strikes_absorbed_per_minute', 'significant_strike_defence', 'average_takedowns_landed_per_15_minutes', 'takedown_accuracy',
       'takedown_defense', 'average_submissions_attempted_per_15_minutes','age']
for col in numeric_cols:                         
    plt.figure(figsize=(6,4))
    sns.histplot(df[col],kde = True,bins = 30)
# made for loop to plot the histogram for each numeric col
```

```python
 sns.boxplot(df['weight_in_kg'],orient='h')
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True),annot=True) #plotted box plot to see outliers & heatmap to see relationships 
```

```python
# fill the null vales in weight by median 
median_weight = df["weight_in_kg"].median()
df['weight_in_kg'] = df['weight_in_kg'].fillna(median_weight)
df.isnull().sum()
```

```python
# from heatmap we get height & reach are very much related so converting weight into categories and filling height and reach with
# category wise median 
df.loc[df["weight_in_kg"] > 200, "weight_in_kg"] = 200 #capping outlier
def weight_group(w):
    if pd.isna(w):
        return np.nan
    elif w < 60:
        return "40-60kg"
    elif w < 90:
        return "60-90kg"
    elif w < 120:
        return "90-120kg"
    else:
        return "120-200kg"

df["weight_group"] = df["weight_in_kg"].apply(weight_group)

# Fill missing height with group median
df["height_cm"] = df.groupby("weight_group")["height_cm"].transform(
    lambda x: x.fillna(x.median())
)

# Fill missing reach with group median
df["reach_in_cm"] = df.groupby("weight_group")["reach_in_cm"].transform(
    lambda x: x.fillna(x.median())
)
```

```python
# From histplot we can see there was many 0s in win,loss etc so making 2 diff datasets for each one so no data is lost 
active_fighters = df[(df['wins'] + df['losses'] + df['draws']) > 0]
new_fighters=df[(df['wins'] + df['losses'] + df['draws']) == 0]
new_fighters.head(10)
```

```python
numeric_cols = ['wins', 'losses', 'draws', 'height_cm','weight_in_kg', 'reach_in_cm',  'significant_strikes_landed_per_minute',
       'significant_striking_accuracy', 'significant_strikes_absorbed_per_minute', 'significant_strike_defence', 'average_takedowns_landed_per_15_minutes', 'takedown_accuracy',
       'takedown_defense', 'average_submissions_attempted_per_15_minutes','age']
for col in numeric_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(active_fighters [col],kde = True,bins = 30)
```

```python
sns.scatterplot(data= active_fighters, x ='age', y ='wins',hue='weight_group', palette= 'muted' ,size='age') #relation between age and wins 
```

```python
# calculating strike to take down ratio to classify the players fighting style 
def calc_ratio(row):
    if row["average_takedowns_landed_per_15_minutes"] == 0:
        return row["significant_strikes_landed_per_minute"] / 0.3  # avoid division by zero so replace it with small no 
    else:
        return row["significant_strikes_landed_per_minute"] / row["average_takedowns_landed_per_15_minutes"]

active_fighters["strike_to_takedown_ratio"] = active_fighters.apply(calc_ratio, axis=1)

#  classify with if/else
def classify(row):
    if row["strike_to_takedown_ratio"] > 3:
        return "Striker"
    else:
        return "Grappler"

active_fighters["Fighting Style"] = active_fighters.apply(classify, axis=1)

#  calculate win percentage
active_fighters["win_percentage"] = active_fighters["wins"] / (active_fighters["wins"] + active_fighters["losses"] + active_fighters["draws"])

# make plot  with seaborn boxplot
plt.figure(figsize=(8,6))
sns.boxplot(x="Fighting Style", y="win_percentage", data=active_fighters)
plt.title("Win Percentage by Fighting Style")
plt.show()
                    
```
